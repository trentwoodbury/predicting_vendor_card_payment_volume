# Imports & Data Formatting

```python
from data_pipeline import data_pipeline

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pda
from scipy.stats import gaussian_kde
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import ElasticNet, LinearRegression
from xgboost import XGBRegressor
```

```python
mpl.rcParams['figure.dpi'] = 175
plt.rcParams['figure.figsize'] = (6,4)

%load_ext autoreload
%autoreload 2
```

```python
pipeline = data_pipeline()
pipeline.perform_full_data_pipeline()

dataset = pipeline.dataset
```

```python
training_split_percent = 0.8
test_split_percent = 1 - training_split_percent

dataset.set_index('organization_id', inplace=True)
eval_columns = ['month_1_cpv', 'month_2_cpv', 'month_3_cpv']
training_columns = [col for col in dataset.columns if col not in eval_columns]

X = dataset.loc[:, training_columns]
month_1 = dataset.loc[:, 'month_1_cpv']
month_2 = dataset.loc[:, 'month_2_cpv']
month_3 = dataset.loc[:, 'month_3_cpv']

row_shuffle = np.random.choice(range(X.shape[0]), replace=False, size=X.shape[0])
cutoff_index = int(training_split_percent * row_shuffle.shape[0])

training_row_indices = row_shuffle[:cutoff_index]
test_row_indices = row_shuffle[cutoff_index:]

X_train = X.iloc[training_row_indices, :].values
m1_train = month_1.iloc[training_row_indices].values
m2_train = month_2.iloc[training_row_indices].values
m3_train = month_3.iloc[training_row_indices].values

X_test = X.iloc[test_row_indices, :].values
m1_test = month_1.iloc[test_row_indices].values
m2_test = month_2.iloc[test_row_indices].values
m3_test = month_3.iloc[test_row_indices].values
```

# Month 1 Modeling

## ElasticNet Regression

```python
# I didn't normalize the data, so we're doing it here.
en = ElasticNet(normalize=True)
en.fit(X_train, m1_train)

predictions = en.predict(X_test)
errors = [np.abs(actual - pred) for pred, actual in zip(m1_test, predictions)]
print("Mean absolute error:", np.average(errors))
```

## Random Forest Regression

```python
rf = RandomForestRegressor(n_estimators=10)
rf.fit(X_train, m1_train)

predictions = rf.predict(X_test)
errors = [np.abs(actual - pred) for pred, actual in zip(m1_test, predictions)]
print("Mean absolute error:", np.average(errors))
```

## Gradient Boosting Regression

```python
gb = GradientBoostingRegressor(n_estimators=10)
gb.fit(X_train, m1_train)

predictions = gb.predict(X_test)
errors = [np.abs(actual - pred) for pred, actual in zip(m1_test, predictions)]
print("Mean absolute error:", np.average(errors))
```

## XGBoost Regression

Let's set the number of estimators very low since the dataset is so small. That way we can avoid overfitting.

```python
xgb = XGBRegressor(objective="reg:squarederror", n_estimators=10)
xgb.fit(X_train, m1_train)

predictions = xgb.predict(X_test)
errors = [np.abs(actual - pred) for pred, actual in zip(m1_test, predictions)]
print("Mean absolute error:", np.average(errors))
```

Let's experiment with adjusting n_estimators.

```python
maes = []
median_errors = []
for ests in np.arange(1,100):
    xgb = XGBRegressor(objective="reg:squarederror", n_estimators=ests)
    xgb.fit(X_train, m1_train)

    predictions = xgb.predict(X_test)
    errors = [np.abs(actual - pred) for pred, actual in zip(m1_test, predictions)]
    maes.append(np.average(errors))
    
plt.plot(np.arange(1,100), maes, label="Mean Absolute Error")
plt.xlabel("n_estimators")
plt.ylabel("Error")
plt.show()
```

## Removing Dummy Variables

```python
X_2 = X.loc[:, ['cpv', 'transaction_count']]
X_train_2 = X_2.iloc[training_row_indices, :].values
X_test_2 = X_2.iloc[test_row_indices, :].values
```

```python
gb = XGBRegressor(objective="reg:squarederror", n_estimators=2)
gb.fit(X_train_2, m1_train)

predictions = gb.predict(X_test_2)
errors = [np.abs(actual - pred) for pred, actual in zip(m1_test, predictions)]
print("Mean absolute error:", np.average(errors))
```

This slightly increases the MAE, so we will include the dummy variables, though they appear to not be contributing very much.

## Removing Outliers

Our big MAE may be coming from outliers, so let's try removing those from the training dataset. I don't know where to set the outlier threshold, so let's just try different thresholds and see which one has the lowest MAE.

```python
thresholds = np.arange(90,100,0.5)
maes = []

for threshold in thresholds:
    X_train_3 = X.iloc[training_row_indices, :]
    X_test_3 = X.iloc[test_row_indices, :].values
    outlier_threshold = np.percentile(X_train_3, threshold)
    inclusion_filter = [True if val < outlier_threshold else False for val in X_train_3['cpv'].values]

    X_train_3 = X_train_3.loc[inclusion_filter, :].values
    y_train_3 = m1_train[inclusion_filter]

    gb = XGBRegressor(objective="reg:squarederror", n_estimators=2)
    gb.fit(X_train_3, y_train_3)

    predictions = gb.predict(X_test_3)
    errors = [np.abs(actual - pred) for pred, actual in zip(m1_test, predictions)]
    maes.append(np.average(errors))
    
plt.plot(thresholds, maes)
plt.title("MAE as a function of outlier cutoff")
plt.ylabel("Mean Absolute Error")
plt.xlabel("Outlier Cutoff Percentile")
plt.show()
```

It looks like the top performance is when we don't remove any outliers.

## Month 1 Model Conclusions

For month 1, our best model is XGBoost setting n_estimators to 2 with no outlier removal.

# Modeling for Month 2

Let's start by using the same parameters for month 2 and compare this to predicting the median for the training period.

```python
gb = XGBRegressor(objective="reg:squarederror", n_estimators=2)
gb.fit(X_train, m2_train)

predictions = gb.predict(X_test)
errors = [np.abs(actual - pred) for pred, actual in zip(m2_test, predictions)]
print(np.average(errors))
```

Let's compare that to predicting the median during the training period.

```python
training_median = np.median(X_train[:, 0]) * (30 / 14)
errors = [np.abs(training_median - actual) for actual in m2_test]
print("Mean absolute error (month 2):", np.mean(errors))
```

It looks like this performance is right about at random, so we need to do some hyperparameter tuning. 

```python
maes = []
median_errors = []
for ests in np.arange(1,100):
    xgb = XGBRegressor(objective="reg:squarederror", n_estimators=ests)
    xgb.fit(X_train, m2_train)

    predictions = xgb.predict(X_test)
    errors = [np.abs(actual - pred) for pred, actual in zip(m2_test, predictions)]
    maes.append(np.average(errors))
    
plt.plot(np.arange(1,100), maes, label="Mean Absolute Error")
plt.xlabel("n_estimators")
plt.ylabel("Error")
plt.show()
```

```python
gb = XGBRegressor(objective="reg:squarederror", n_estimators=1)
gb.fit(X_train, m2_train)

predictions = gb.predict(X_test)
errors = [np.abs(actual - pred) for pred, actual in zip(m2_test, predictions)]
print(np.average(errors))
```

Since using a single estimator maximizes performance, let's simplify and only look at CPV as an input feature, seeing if that increases performance a bit more.

```python
gb = XGBRegressor(objective="reg:squarederror", n_estimators=1)
gb.fit(X_train[:, 0].reshape(-1,1), m2_train)

predictions = gb.predict(X_test[:, 0].reshape(-1,1))
errors = [np.abs(actual - pred) for pred, actual in zip(m2_test, predictions)]
print(np.average(errors))
```

This increases the MAE slightly. What if we remove the dummy variables?
